```python
import pandas as pd
```

```python
df = pd.read_csv('billboard.csv')
df
```

```python
billboard = pd.melt(df, id_vars=['artiste', 'track', 'genre', 'date.entered', 'date.peaked'], 
               value_name='rank', var_name='week')
billboard
```

```python
billboard['week'] = billboard['week'].str.extract('(\d+)').astype(int)
billboard.head(5)
```

```python
billboard['date.entered'] = pd.to_datetime(billboard['date.entered'])
```

```python
billboard['date.peaked'] = pd.to_datetime(billboard['date.peaked'])
```

```python
billboard['entered_to_peak'] = billboard['date.peaked'] - billboard['date.entered']
billboard
```

```python
billboard['entered_to_peak'] = billboard['entered_to_peak'].dt.days
billboard
```

```python
billboard = billboard.dropna(thresh=8)
billboard.head(6)
```

```python
billboard.dtypes
```

```python
billboard.shape
```

```python
billboard = billboard[['artiste', 'track', 'rank', 'genre', 'entered_to_peak']][billboard['entered_to_peak'] == 35]
billboard
```

```python
billboard[['artiste', 'track', 'rank', 'genre', 'entered_to_peak']][billboard['genre'] == 'Rock'].sort_values('rank')
```
